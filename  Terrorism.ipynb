```python
import pandas as pd
```

```python
df = pd.read_csv('terrorism.csv' , encoding = 'latin1', low_memory=False)
df.head(5)
```

```python
df.columns.tolist()
```

```python
newdf = df[[ 
 'iyear',
 'imonth',
 'iday',
 'country_txt',
'region_txt',
 'provstate',
  'city',
 'latitude',
 'longitude',
  'summary',
'attacktype1_txt', 
'weapsubtype1_txt',
'scite1',
 'scite2',
 'scite3',
 
  ]]

newdf.head(3)
```

```python
cleandf1 = newdf.dropna()
cleandf1.head(2)
```

```python
 cleandf1['imonth'].unique()
```

```python
pd.options.mode.copy_on_write = True
month_name = pd.to_datetime(cleandf1['imonth'], format='%m').dt.month_name()
month_name
```

```python
cleandf1[cleandf1['iday'] == 0].head(3)
```

```python
splited = cleandf1["summary"].str.split(r":", expand = True, n=1).rename(columns = {0:"dates", 1:"info"})
splited
```

```python
splited2 = splited["dates"].str.split(r"(", expand = True, n=1).rename(columns = {0:"date", 1:"y"})
splited2
 
```

```python
splited2['date'].unique()
```

```python

splited2['new_dy'] = splited2['date'].str[2:5]
splited2['new_dy']
```

```python
new_day1 = splited2['new_dy'].str.lstrip('/')
new_day1
```

```python
new_day2 = new_day1.str.split(r"/", expand = True, n=1).rename(columns = {0:"day", 1:"nyd"})

new_day2
```

```python
new_day2['day'].unique()
```

```python
cleandf2 = pd.concat([ cleandf1, splited, splited2, new_day2], axis=1)
cleandf2.head(3)
```

```python
#droped = cleandf2.drop(columns=['iday','summary','dates','new_dy','y','nyd'], inplace=True)
#droped
cleandf2.head(3)
```

```python
vl = cleandf2[(cleandf2['day'] == '00') | (cleandf2['day'] == '0') | (cleandf2['day'] == ' Th')]
vl.head(3)
```

```python

vlu = (cleandf2['day'] == '00') | (cleandf2['day'] == '0') | (cleandf2['day'] == ' Th')
vlu.value_counts()
```

```python
cleandf3 =  cleandf2[(cleandf2['day'] != '00') | (cleandf2['day'] != '0') | (cleandf2['day'] != ' Th')].copy()
cleandf3.head()
 
```
